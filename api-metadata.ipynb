```python
import requests
import pandas as pd
import json
import time

url1 = 'https://raw.githubusercontent.com/e3la/eda-checkpoint/starter-main/articles.csv'
fac = pd.read_csv(url1) #this is the faculty provided dataset

print(fac['Title'].head())
print(fac.info())
title_list = fac['Title'].iloc[4600:]
print(title_list)
```

```python
df = fac
row_index = df[df['Title'].str.contains('The paradox of tolerance')].index[0]

print(row_index)
```

```python
#calling the API with headers and a for loop to go through title list

API_KEY = '6a2a34c6-7fb4-4bb6-a1e3-5221ba28cabf'

headers = {
    'api_key': API_KEY,
    'Content-Type': 'application/json'
}

# Initialize an empty DataFrame
result_df = pd.DataFrame()

for title in title_list:
    resp = requests.get('https://bg.api.oa.works/metadata?id=' + title, headers=headers)

    # Check if the request was successful
    if resp.status_code == 200:
        data = resp.json()  # Parse the JSON response

        # Create a DataFrame from the JSON data, handling missing keys with None values
        df = pd.DataFrame([data], columns=data.keys())

        # Append the new data to the existing DataFrame
        result_df = result_df.append(df, ignore_index=True)
    else:
        print('Failed to retrieve data for Title:', title)

    # Introduce a delay of 1 second between API requests
    time.sleep(1)

# Print the final DataFrame with data from all API requests
print(result_df.head())
```
